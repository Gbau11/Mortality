```python
muertes_por_pais_y_anio = df.groupby(['Country', 'Year'])['Total'].sum()

# Calcular la probabilidad de morir por país y año
probabilidad_muerte_por_pais_y_anio = muertes_por_pais_y_anio / df.groupby(['Country', 'Year'])['Population'].mean()

# Calcular la probabilidad media de morir por país
probabilidad_media_muerte_por_pais = probabilidad_muerte_por_pais_y_anio.groupby('Country').mean()

# Imprimir la probabilidad media de morir por país
probabilidad_media_muerte_por_pais.sort_values(ascending=False)
```

```python
df_1991 = df[df['Year'] == 1991]
df_2016 = df[df['Year'] == 2016]
differences = []  # Lista para almacenar las diferencias

for c in df_2016['Country']:
    obesity_country_2016 = df_2016.loc[df_2016['Country'] == c, 'Obesity'].values
    obesity_country_1991 = df_1991.loc[df_1991['Country'] == c, 'Obesity'].values
    if len(obesity_country_2016) > 0 and len(obesity_country_1991) > 0:
        diff = obesity_country_2016[0] - obesity_country_1991[0]
        differences.append((c, diff))  # Añadir el país y la diferencia a la lista
    else:
        print("No se encontraron datos para", c)

# Ordenar la lista de diferencias de mayor a menor
differences_sorted = sorted(differences, key=lambda x: x[1], reverse=True)

# Imprimir los resultados ordenados
for country, diff in differences_sorted:
    print(country, diff)
```
