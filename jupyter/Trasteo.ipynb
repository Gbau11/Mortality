```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 150)
from scipy.stats import spearmanr, pearsonr
```

```python
df = pd.read_csv('../csv/completa_per_cap.csv')
```

```python
df
```

```python
muertes_por_pais_y_anio = df.groupby(['Country', 'Year'])['Total'].sum()

# Calcular la probabilidad de morir por paÃ­s y aÃ±o
probabilidad_muerte_por_pais_y_anio = muertes_por_pais_y_anio / df.groupby(['Country', 'Year'])['Population'].mean()

# Calcular la probabilidad media de morir por paÃ­s
probabilidad_media_muerte_por_pais = probabilidad_muerte_por_pais_y_anio.groupby('Country').mean()

# Imprimir la probabilidad media de morir por paÃ­s
probabilidad_media_muerte_por_pais.sort_values(ascending=False)
```

```python
lista_correlaciones = []

for a in df['Year'].unique():

    datos_a = df[df['Year'] == a]
    
    correlation, p_value = pearsonr(datos_a['Neoplasms/cap'], datos_a['Obesity(%)'])
    
    lista_correlaciones.append((a, correlation, p_value))

#lista_correlaciones

tuplas_significativas = [(a, correlacion, p_valor) for a, correlacion, p_valor in lista_correlaciones if p_valor < 0.05]

cantidad_significativas = len(tuplas_significativas)

porcentaje_significativas = (cantidad_significativas / len(lista_correlaciones)) * 100

print("Porcentaje de tuplas significativas respecto al total:", porcentaje_significativas, "%")

tuplas_significativas
```

```python
lista_correlaciones = []

for pais in df['Country'].unique():

    datos_pais = df[df['Country'] == pais]
    
    correlation, p_value = pearsonr(datos_pais['Neoplasms/cap'], datos_pais['Obesity(%)'])
    
    lista_correlaciones.append((pais, correlation, p_value))

#lista_correlaciones

tuplas_significativas = [(pais, correlacion, p_valor) for pais, correlacion, p_valor in lista_correlaciones if p_valor < 0.05]

cantidad_significativas = len(tuplas_significativas)

porcentaje_significativas = (cantidad_significativas / len(lista_correlaciones)) * 100

print("Porcentaje de tuplas significativas respecto al total:", porcentaje_significativas, "%")

tuplas_significativas
```
