```python
df['Alzheimers Disease and Other Dementias/cap'] = df['Alzheimers Disease and Other Dementias']/df['Population']
df['Acute Hepatitis/cap'] = df['Acute Hepatitis']/df['Population']
df['Digestive Diseases/cap'] = df['Digestive Diseases']/df['Population']
df['Cirrhosis and Other Chronic Liver Diseases/cap'] = df['Cirrhosis and Other Chronic Liver Diseases']/df['Population']
df['Poisonings/cap'] = df['Poisonings']/df['Population']
df['Road Injuries/cap'] = df['Road Injuries']/df['Population']
df['Conflict and Terrorism/cap'] = df['Conflict and Terrorism']/df['Population']
df['Environmental Heat and Cold Exposure/cap'] = df['Environmental Heat and Cold Exposure']/df['Population']
df['Self-harm/cap'] = df['Self-harm']/df['Population']
df['Alcohol Use Disorders/cap'] = df['Alcohol Use Disorders']/df['Population']
df['HIV/AIDS/cap'] = df['HIV/AIDS']/df['Population']
df['Nutritional Deficiencies/cap'] = df['Nutritional Deficiencies']/df['Population']
df['Interpersonal Violence/cap'] = df['Interpersonal Violence']/df['Population']
df['Neoplasms/cap'] = df['Neoplasms']/df['Population']
df['Neonatal Disorders/cap'] = df['Neonatal Disorders']/df['Population']
df['Respiratory Diseases/cap'] = (df['Lower Respiratory Infections']+df['Chronic Respiratory Diseases'])/df['Population']
df['Tuberculosis/cap'] = df['Tuberculosis']/df['Population']
```

```python
df.sum(axis=0,numeric_only=True).sort_values(ascending=False)
```
